```python
import matplotlib.pyplot as plt

x = np.arange(0, 0.4, 0.01)
print(len(x))
y1 = ad_dis1/10
y2 = ad_dis2/10
y3 = ad_dis3/10

#y4 = (y1 + y2 + y3)/3

y4 = ad_dis4/10


z1 = np.polyfit(x, y4, 1)#用3次多项式拟合
p1 = np.poly1d(z1)
print(p1) #在屏幕上打印拟合多项式
newx = np.arange(0, 0.4, 0.01)
yvals=p1(newx)#也可以使用yvals=np.polyval(z1,x)
#plot1=plt.plot(x, y1, '*',label='ad1')
#plot2=plt.plot(x, y2, '.',label='ad2')
#plot3=plt.plot(x, y3, '+',label='ad3')
plot4=plt.plot(x, y4, '^',label='ada')



plot5=plt.plot(newx, yvals, 'r',label='polyfit values')

plot6 = plt.plot(x, y1 - 273.7*x - 1.066, 'r',label='polyfit values2')

plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend(loc=4)
plt.title('polyfitting')
plt.show()

```
