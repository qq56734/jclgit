```python
def segmentation(partition):
    import os
    import re
    import jieba
    import jieba.analyse
    import jieba.posseg as pseg
    import codecs

    abspath = "C:/Users/fscut/Desktop/OA讨论"

    # # 结巴加载用户词典
    #userDict_path = os.path.join(abspath, "ITKeywords.txt")
    #jieba.load_userdict(userDict_path)
    #
    # # 停用词文本
    stopwords_path = os.path.join(abspath, "stopwords.txt")
    def get_stopwords_list():
    #     """返回stopwords列表"""
         stopwords_list = [i.strip() for i in codecs.open(stopwords_path, encoding = 'UTF-8').readlines()]
         return stopwords_list
    # # 所有的停用词列表
    stopwords_list = get_stopwords_list()



    # 分词
    def cut_sentence(sentence):
        """对切割之后的词语进行过滤，去除停用词，保留名词，英文和自定义词库中的词，长度大于2的词"""
        seg_list = pseg.lcut(sentence)
        seg_list = [i for i in seg_list if i.word not in stopwords_list]
        filtered_words_list = []
        for seg in seg_list:
            if len(seg.word) <= 1:
                continue
            elif seg.flag == "eng":
                if len(seg.word) <= 2:
                    continue
                else:
                    filtered_words_list.append(seg.word)
            elif seg.flag.startswith("n"):
                filtered_words_list.append(seg.word)
            elif seg.flag in ["eng"]:  # 是自定一个词语或者是英文单词
                filtered_words_list.append(seg.word)
        return filtered_words_list

    result = []
    for row in partition:
        #sentence = re.sub("<.*?>", "")  # 替换掉标签数据
        words = cut_sentence(row)
        result += words
    return result
```

```python
import os

article_dicts = []

base_dir = r'C:\Users\fscut\Desktop\OA讨论\reports'

for id in ids[1:] :

    print(id)

    filename = id


    path = os.path.join(base_dir, id + '.txt')

    with open(path, "r", encoding='utf-8') as f:
        title = f.readline().split()
        article = f.read().split()

    article = article + title

    seg = segmentation(article)



    article_dict  = {}
    article_dict['id'] = id
    article_dict['title'] = ''.join(title)
    article_dict['words'] = seg


    article_dicts.append(article_dict)
```

```python
from pyspark.sql import SparkSession
from pyspark.conf import SparkConf
from pyspark.sql.types import *
from pyspark.sql import functions as F, Window
import pandas


config = SparkConf()
# config.set('spark.dynamicAllocation.maxExecutors', '8')
# config.set('spark.driver.memory', '4G')
# config.set('spark.executor.memory', '8G')
# config.set('spark.executor.cores', '8')
# config.set('spark.yarn.executora.memoryOverhead', '4G')
# config.set('spark.sql.shuffle.partitions', '500')
# config.set('spark.default.parallelism', '500')
# config.set('spark.port.maxRetries', '1000')
# config.set('spark.sql.sources.partitionOverwriteMode', 'dynamic')
config.set('spark.master','local[4]')

spark = SparkSession.builder.config(conf=config).getOrCreate()



#spark = SparkSession.builder.master('local').appName("test_script").getOrCreate()

#article_df  = spark.createDataFrame( article_dict)
words_df = spark.createDataFrame(article_dicts)
print(words_df.take(1))
```

```python
from pyspark.ml.feature import Word2Vec
from pyspark.ml.feature import Word2VecModel

w2v_model = Word2Vec(vectorSize=100, inputCol='words', outputCol='vector', minCount=3)
model = w2v_model.fit(words_df)

model.write().overwrite().save("C:/Users/fscut/Desktop/OA讨论/word2vec_model/python.word2vec")

from pyspark.ml.feature import Word2VecModel

w2v_model = Word2VecModel.load("C:/Users/fscut/Desktop/OA讨论/word2vec_model/python.word2vec")
vectors = w2v_model.getVectors()
vectors.show()
```

```python
from pyspark.ml.feature import CountVectorizer

# vocabSize是总词汇的大小，minDF是文本中出现的最少次数
cv = CountVectorizer(inputCol="words", outputCol="countFeatures", vocabSize=200 * 10000, minDF=1.0)
# 训练词频统计模型
cv_model = cv.fit(words_df)
cv_model.write().overwrite().save("C:/Users/fscut/Desktop/OA讨论/models/CV.model")

from pyspark.ml.feature import CountVectorizerModel

cv_model = CountVectorizerModel.load("C:/Users/fscut/Desktop/OA讨论/models/CV.model")
# 得出词频向量结果
cv_result = cv_model.transform(words_df)

# idf
from pyspark.ml.feature import IDF

idf = IDF(inputCol="countFeatures", outputCol="idfFeatures")
idf_model = idf.fit(cv_result)
idf_model.write().overwrite().save("C:/Users/fscut/Desktop/OA讨论/models/IDF.model")

# tf-idf
from pyspark.ml.feature import IDFModel

idf_model = IDFModel.load("C:/Users/fscut/Desktop/OA讨论/models/IDF.model")
tfidf_result = idf_model.transform(cv_result)

# 选取前20个作为关键词,此处仅为词索引
def sort_by_tfidf(partition):
    TOPK = 20
    for row in partition:
        # 找到索引与IDF值并进行排序
        _dict = list(zip(row.idfFeatures.indices, row.idfFeatures.values))
        _dict = sorted(_dict, key=lambda x: x[1], reverse=True)
        result = _dict[:TOPK]
        for word_index, tfidf in result:
            yield row.id, int(word_index), round(float(tfidf), 4)

keywords_by_tfidf = tfidf_result.rdd.mapPartitions(sort_by_tfidf).toDF(["id", "index", "weights"])

# 构建关键词与索引
keywords_list_with_idf = list(zip(cv_model.vocabulary, idf_model.idf.toArray()))

def append_index(data):
    for index in range(len(data)):
        data[index] = list(data[index])  # 将元组转为list
        data[index].append(index)  # 加入索引
        data[index][1] = float(data[index][1])

append_index(keywords_list_with_idf)
sc = spark.sparkContext
rdd = sc.parallelize(keywords_list_with_idf)  # 创建rdd
idf_keywords = rdd.toDF(["keywords", "idf", "index"])

# 求出文章关键词及权重tfidf
keywords_result = keywords_by_tfidf.join(idf_keywords, idf_keywords.index == keywords_by_tfidf.index).select(
    ["id", "keywords", "weights"])
print("关键词权重", keywords_result.take(10))
#keywords_result.write.mode('overwrite').format("csv").save("C:/Users/fscut/Desktop/OA讨论/关键词权重")

# 文章关键词与词向量join
keywords_vector = keywords_result.join(vectors, vectors.word == keywords_result.keywords, 'inner')
```

```python
from sqlalchemy import create_engine, Integer, VARCHAR
from util import consts
import sqlalchemy

engine = create_engine(consts.MYSQL_CONNECTION_BIGDATA_ARTICLE)
connect = engine.connect()

keywords_result.toPandas().to_sql(name="article_keywords_weights", con=connect, if_exists="append", index=False,
                     dtype={'article_id': Integer(),
                            'keyword': VARCHAR(
                                length=255),
                            "weights" : sqlalchemy.types.Float
                          
                            })
```

```python
def compute_vector(row):
    return row.id, row.keywords, row.weights * row.vector

article_keyword_vectors = keywords_vector.rdd.map(compute_vector).toDF(["id", "keywords", "weightingVector"])


# 利用 collect_set() 方法，将一篇文章内所有关键词的词向量合并为一个列表
article_keyword_vectors.registerTempTable('temptable')
article_keyword_vectors = spark.sql("select id, collect_set(weightingVector) vectors from temptable group by id")
```

```python
article_keyword_vectors.write.mode('overwrite').format("json").save("C:/Users/fscut/Desktop/OA讨论/article_keyword_vectors.json")
```

```python
print("article vector go")
article_keyword_vectors.show()
def compute_avg_vectors(row):
    x = 0
    for i in row.vectors:
        x += i
    # 求平均值
    return row.id, x / len(row.vectors)

article_vector = article_keyword_vectors.rdd.map(compute_avg_vectors).toDF(['id', 'articlevector'])
print("文章最终vector",article_vector.take(2))
```
