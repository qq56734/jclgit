```python
import os

articles = []

base_dir = r'C:\Users\fscut\Desktop\OA讨论\reports'
for id in ids[1:] :
    print(id)
    path = os.path.join(base_dir, id + '.txt')
    with open(path, "r", encoding='utf-8') as f:
        title = f.readline().split()
        article = f.read()
        article_dict  = {}
        article_dict['id'] = id
        article_dict['title'] = ''.join(title)
        article_dict['content'] = article
        articles.append(article_dict)
```

```python
article_df = pandas.DataFrame(articles)
article_df
```
